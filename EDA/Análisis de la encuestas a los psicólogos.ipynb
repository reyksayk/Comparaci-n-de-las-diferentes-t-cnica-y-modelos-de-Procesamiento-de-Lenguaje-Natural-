```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
A = '/content/drive/MyDrive/Experimentación Python Tesis /encuesta a psicologos/Encuesta a Psicólogos_ Sintomatología asociada a la salud mental en las redes sociales(1-14).xlsx'
data = pd.read_excel(A)
data.info()
data.head()
```

```python
columna_formacion_academica = data['¿Cuál es su formación académica en el área de salud mental? (Seleccione todas las opciones que apliquen)']

# Dividir las palabras y convertir la serie resultante en una lista de listas
formacion_academica_lista = columna_formacion_academica.str.split(';').tolist()

# Crear un DataFrame a partir de la lista de listas
df= pd.DataFrame(formacion_academica_lista, columns=[f'Palabra_{i+1}' for i in range(max(map(len, formacion_academica_lista)))])

# Unir las columnas en una única columna 'Palabra' y contar la frecuencia
df = df.melt(value_name='Palabra').dropna()['Palabra'].value_counts().reset_index()
df.columns = ['Formación Académica', 'Frecuencia']
df = df.drop(0)
df
```

```python
# Crear el gráfico de torta por porcentaje
fig = px.pie(df,
             names='Formación Académica',
             values='Frecuencia',
             title='Distribución de Formación Académica en Salud Mental',
             labels={'Frecuencia': 'Porcentaje'},
             width=1400, height=600)

# Mostrar el gráfico
fig.show()
```

```python
data['¿Cuántos años de experiencia tiene en el campo de la salud mental?'] = data['¿Cuántos años de experiencia tiene en el campo de la salud mental?'].str.strip()

Df= data['¿Cuántos años de experiencia tiene en el campo de la salud mental?'].unique()
Df = data['¿Cuántos años de experiencia tiene en el campo de la salud mental?'].value_counts().reset_index()
Df.columns = ['Años de experiencia', 'Frecuencia']
Df.head()
```

```python
columna_formacion_academica = data['¿En cuáles de las siguientes redes sociales ha observado publicaciones relacionadas con problemas de depresión y ansiedad? (Seleccione todas las opciones que apliquen)']

# Dividir las palabras y convertir la serie resultante en una lista de listas
formacion_academica_lista = columna_formacion_academica.str.split(';').tolist()

# Crear un DataFrame a partir de la lista de listas
df= pd.DataFrame(formacion_academica_lista, columns=[f'Palabra_{i+1}' for i in range(max(map(len, formacion_academica_lista)))])

# Unir las columnas en una única columna 'Palabra' y contar la frecuencia
df = df.melt(value_name='Palabra').dropna()['Palabra'].value_counts().reset_index()
df.columns = ['Redes sociales', 'Frecuencia']
df = df.drop(0)
df
```

```python
columna_formacion_academica = data['Seleccione las palabras más comunes que las personas con depresión utilizan en sus publicaciones en redes sociales: (Seleccione todas las opciones que apliquen)']

# Dividir las palabras y convertir la serie resultante en una lista de listas
formacion_academica_lista = columna_formacion_academica.str.split(';').tolist()

# Crear un DataFrame a partir de la lista de listas
df= pd.DataFrame(formacion_academica_lista, columns=[f'Palabra_{i+1}' for i in range(max(map(len, formacion_academica_lista)))])

# Unir las columnas en una única columna 'Palabra' y contar la frecuencia
df = df.melt(value_name='Palabra').dropna()['Palabra'].value_counts().reset_index()
df.columns = ['palabras', 'Frecuencia']
df = df.drop(index=[0, 12])
df
```

```python
import plotly.express as px

# Crear el gráfico de barras con frecuencia mostrada en las barras
fig = px.bar(df, x='palabras', y='Frecuencia',
             title='Palabras más usadas en Redes Sociales según los Psicólogos (depresión)',
             labels={'Frecuencia': 'Frecuencia'},
             width=1000, height=600)

# Ajustes visuales
fig.update_traces(marker_color='skyblue', marker_line_color='black', marker_line_width=1, opacity=0.7,
                  text=df['Frecuencia'])  # Mostrar la frecuencia en las barras

fig.update_layout(xaxis_tickangle=-45, xaxis=dict(type='category'))

# Ajustes de diseño
fig.update_layout(
    margin=dict(l=50, r=50, t=80, b=50),  # Márgenes para mejorar la visualización
    showlegend=False  # No mostrar la leyenda
)

# Mostrar el gráfico
fig.show()
```
